```python
# exctractingCoursesNames function exctract the courses names to new list to help in the filtering part
def exctractingCoursesNames(list):
    n = len(list)
    newList = []
    for i in range(0,n):
        if i == 0 :
            newList.append(list[i])
            print(list[i], 'item added')
        elif i > 0 :
            if list[i] == list[i-1]:
                print(list[i], 'item is allready exist.')
                continue
            else:
                     newList.append(list[i]) 
                     print(list[i], 'item added')
    return  newList         
   
            
coursesList = exctractingCoursesNames(nameList)   
print('\nCourses: \n',coursesList)
```
