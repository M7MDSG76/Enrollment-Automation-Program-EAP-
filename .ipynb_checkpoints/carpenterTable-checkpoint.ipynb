```python
# step 7
# exctractingCoursesNames function exctract the courses names to nameList to help in the filtering part

def exctractingCoursesNames(list):
    n = len(list)
    newList = []
    for i in range(0,n):
        if i == 0 :
            newList.append(list[i])
            
        elif i > 0 :
            if list[i] == list[i-1]:
               
                continue
            else:
                     newList.append(list[i]) 
                     
    return  newList         
        
coursesList = exctractingCoursesNames(nameList)   
coursesList
```
